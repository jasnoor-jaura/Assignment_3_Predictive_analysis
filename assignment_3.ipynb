```python
# # Learn Probability Density Functions using Roll-Number Parameterized Non-Linear Model

# This assignment estimates parameters of a probability density function after applying a non-linear transformation on air quality data.
# Feature used: NOâ‚‚ concentration
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("/content/data.csv.zip", encoding='latin1')

col=[c for c in data.columns if "no2" in c.lower() or "no_2" in c.lower()]
x=data[col[0]].dropna().values

z=x+0.15*np.sin(0.3*x)

mu=np.mean(z)
sigma2=np.var(z)
lam=1/(2*sigma2)
c=1/np.sqrt(2*np.pi*sigma2)

print(mu,lam,c)

plt.hist(z,bins=50,density=True,alpha=0.5)
z_vals=np.linspace(min(z),max(z),500)
pdf=c*np.exp(-lam*(z_vals-mu)**2)
plt.plot(z_vals,pdf,'r')
plt.show()
```
